```python
#casefolding
import re

def casefolding(text):
  text = text.lower()
  text = re.sub(r'https?://\S+|www\.\S+','',text)
  text = re.sub(r'[-+]?[0-9]+','',text)
  text = re.sub(r'[^\w\s]','', text)
  text = text.strip()
  return text

```

```python
raw_Sample = data['Tweet'].iloc[4]
case_folding = casefolding(raw_Sample)

print('Raw data\t :', raw_Sample)
print('case folding\t :', case_folding)
```

```python
key_norm=pd.read_excel('kamus perbaikan kata.xlsx')

def tex_normalize(text):
  text = ' '.join([key_norm[key_norm['TIDAK BAKU']== word]['BAKU'].values[0]
  if (key_norm['TIDAK BAKU']== word).any()
  else word for word in text.split()               
  ])

  text = str.lower(text)
  return text
```

```python
raw_data = data['Tweet'].iloc[696]
word_normal = tex_normalize(case_folding)

print('Raw_Data\t :', raw_data)
print('word normalize\t :', word_normal)


```

```python
# stopword removal
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

stopwords_ind = stopwords.words('indonesian')
```

```python
len(stopwords_ind)
```

```python
#menambahkan kata dalam stopwords
more_stopwords = ['bulutangkis','grey','polii','wkwkwkwk', 'masters','istora', 'malaysia','nonton']
stopwords_ind = stopwords_ind + more_stopwords

def remove_stop_word(text):
  clean_words = []
  text = text.split()
  for word in text:
    if word not in stopwords_ind:
      clean_words.append(word)
  return " ".join(clean_words)
```

```python
raw_sample = data['Tweet'].iloc[696]
case_folding = casefolding(raw_sample)
stopwords_removal = remove_stop_word(case_folding)

print('raw data\t\t :', raw_sample)
print('case Folding\t\t :', case_folding)
print('stopword\t\t', stopwords_removal)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()

stemmer = factory.create_stemmer()

#fungsi

def stemming(text):
  text = stemmer.stem(text)
  return text

```

```python
# gabung fungsi
def text_preprocessing_process(text):
  text = casefolding(text)
  text = tex_normalize(text)
  text = remove_stop_word(text)
  text = stemming(text)
  return text
```

```python
%%time
data['clean_teks'] = data['Tweet'].apply(text_preprocessing_process)
```

```python
# Feature engineering
X = data['clean_teks']
Y = data['Target']
```

```python
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer

vec_TF_IDF = TfidfVectorizer(ngram_range=(1,1))
vec_TF_IDF.fit(X)

x_tf_idf = vec_TF_IDF.transform(X)

pickle.dump(vec_TF_IDF.vocabulary_,open('fiture_tf_idf.sav', 'wb'))
```

```python
vec_TF_IDF.vocabulary_
```

```python
#jumlah fitur
print(len(vec_TF_IDF.get_feature_names_out()))
```

```python
X1=vec_TF_IDF.transform(X).toarray()
data_tfidf=pd.DataFrame(X1,columns=vec_TF_IDF.get_feature_names_out())
data_tfidf.head(5)
```

```python
x_train = np.array(data_tfidf)
y_train = np.array(Y)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

Feature_Chi2=SelectKBest(chi2, k=2500)
x_kbest_feature=Feature_Chi2.fit_transform(x_train, y_train)

# Reduced feature
print('Original feature number', x_train.shape[1])
print('Reduced feature number', x_kbest_feature.shape[1])
```

```python
data3=pd.DataFrame(Feature_Chi2.scores_, columns=["Nilai"])
data3.head(5)
```

```python
feature=vec_TF_IDF.get_feature_names_out()
feature

data3['Fitur']=feature
data3.head(5)
```

```python
data3.sort_values(by='Nilai', ascending=False)
```

```python
mask=Feature_Chi2.get_support()
mask
```

```python
fitur_baru = []
for bool, f in zip(mask, feature):
  if bool :
    fitur_baru.append(f)
  selected_feature = fitur_baru
selected_feature
```

```python
# buat vocabulary baru
fitur_baru_seleksi = {}

for (k,v) in vec_TF_IDF.vocabulary_.items():
  if k in selected_feature:
    fitur_baru_seleksi[k]=v

fitur_baru_seleksi
```

```python
import random
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
Algoritma = MultinomialNB()
```

```python
model = Algoritma.fit(X_train, Y_train)
```
